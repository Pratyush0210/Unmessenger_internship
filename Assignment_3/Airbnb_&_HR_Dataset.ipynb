#### **Barplot of Average price v/s Room type**
- Plotly module is used to show interactive barchart.
- Entire room/apt is having highest average price of $188.57.

```python
air = airbnb.groupby("room_type").agg({"price": 'mean'})
px.bar( air, text_auto='', color= list(air.T.columns), title= 'Average Price w.r.t. Room Type', width=700, height=525)
```

#### **Barplot of Average price v/s neighbourhood_group**
- Plotly module is used to show interactive barchart.
- Manhattan is having highest average price of $151.84.

- Bronx is having least average price of $42.5.

```python
areas = airbnb.groupby("neighbourhood_group").agg({"price": 'mean'})
px.bar( areas, text_auto='', color= list(areas.T.columns), title= 'Average Price w.r.t. Neighbourhood Group', width=1000, height=525)
```

## **MULTI-VARIATE ANALYSIS**

#### **Average price with respect to room type in every neighbourhood group**
- Average price of Entire home/apt in Queens is highest.
- Entire home/apt are not present in Staten island and Bronx.
- Shared rooms are only in Manhattan with Average price of approx. $75.

```python
plt.figure(figsize =(15,7))
sns.barplot(data = airbnb, x= 'neighbourhood_group', y= 'price', hue='room_type', errorbar= None)
```

#### **Scatterplot between Longitude and Latitude with distinct colors of Neighbourhood group**
- Red dots are more than the other colors dots, which shows, Manhattan has highest number of Airbnb hostels.
- this Scatterplot represent the map of New York.
- There is only one purple dot which shows least number of hostel present in Staten Island.

```python

fig1 = px.scatter(airbnb, x= 'longitude', y= 'latitude', color= 'neighbourhood_group', hover_data=['neighbourhood'])
fig1.update_traces(marker_size=10)
fig1.show()
```

#### **Scatterplot between availability_365 and price with continuous colour shades of price**
- Highest rate of hostel available in Brooklyn Heights for 7 days per year of $800. 
- this Scatterplot represent availability of hostels in days per year to the price of hostels.
- All columns are available like name of hostel, host name, neighbourhood, neighbourhood group, with the price and no. of days availability per year.

```python
fig = px.scatter(airbnb, x="availability_365", y="price", color='price', hover_data=['name', 'host_name', 'neighbourhood', 'neighbourhood_group'])
fig.update_traces(marker_size=13)
fig.show()
```
